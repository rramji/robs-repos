```python
def get_connections(molecule):
    cutoff = neighborlist.natural_cutoffs(molecule)
    nl = neighborlist.NeighborList(cutoff, self_interaction=False, bothways=True)
    nl.update(molecule)
    matrix = nl.get_connectivity_matrix()
    return matrix
```

```python
bond_matrix = get_connections(mol).toarray()
```
